```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
# iris=load_iris()
# X=iris.data
# Y=iris.target

# df=pd.read_csv('diabetes.csv')
# X=df.iloc[:,:-1].values
# Y=df.iloc[:,-1:].values

df1=pd.read_csv("C:/titanic.csv")
X=df1.drop(columns=["Name","Sex","Survived"]).values
Y=df1["Survived"].values

K=3
centriods=X[np.random.choice(X.shape[0],K,replace=False)]
previous_centriods=np.zeros_like(centriods)
clusters=np.zeros(len(X))
```

```python
def euclidian(x,x1):
    return np.sqrt(np.sum(np.square(x-x1)))
while not np.array_equal(previous_centriods,centriods):
    previous_centriods=centriods.copy()
    
    for i,x in enumerate(X):
        distances=[euclidian(x,centriod) for centriod in centriods]
        print(distances)
        cluster=np.argmin(distances)
        print(cluster)
        clusters[i]=cluster
        
    for k in range(K):
        cluster_points=[X[i] for i in range(len(X)) if cluster[i]==k]
        centroids[k]=np.mean(cluster_points,axis=0)
        
```

```python
kmeans = KMeans(n_clusters=K)

# Fit the KMeans model to the data
kmeans.fit(X)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Get the cluster labels for each data point
labels = kmeans.labels_

# Plot the clustered data points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', alpha=0.5)

# Plot the cluster centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering of Iris Dataset')
plt.show()
```
